### 15.Calculate and display

Make a dataframe with 3 columns that holds:
- all city names
- how many people live in each city
- average age for each city 

Tips:
- groupby
- count
- mean
- do this in several steps

```python
residents_per_city = df.groupby('City').Name.count()

df_display = df[['City']]
df_display.insert(1, "Population", residents_per_city, True)
df_display.insert(2, "Average age", df.Age.mean(), True)

df_display
```
